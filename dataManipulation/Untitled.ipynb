```python
import pandas as pd
```

```python
url = "https://www.weatherandclimate.info/monitor/?id=42807&month=8&year=2024"
```

```python
temperatures = pd.read_html(url)
```

```python
sepTemps = pd.DataFrame(temperatures[0])
```

```python
columns1 = sepTemps.iloc[0, :].to_list()
```

```python
columns2 = sepTemps.iloc[1, :].to_list()
```

```python
columns_rough = []
for i in range(len(columns2)):
    if columns1[i] != columns2[i]:
        columns_rough.append(columns1[i] + " " + columns2[i])
    else:
        columns_rough.append(columns1[i])
```

```python
final_rows_set = sepTemps.drop(index = [0, 1])
```

```python
final_set = final_rows_set.set_axis(columns_rough, axis=1)
```
