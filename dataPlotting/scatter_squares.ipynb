```python
import matplotlib.pyplot as plt

# argument to set the size of the dots used to draw the graph.
# plt.scatter(2, 4, s = 50)
# x_values = [1, 2, 3, 4, 5]
# y_values = [1, 4, 9, 16, 25]
# plt.scatter(x_values, y_values, s = 40)
x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, s = 10, c = y_values, cmap = plt.cm.Paired, edgecolor='none')

# Set chart title and label axes.
plt.title("Range of Square Numbers", fontsize = 12)
plt.xlabel("value", fontsize = 9)
plt.ylabel("square of value", fontsize = 9)

# Set size of tick labels.
plt.tick_params(axis='both', labelsize=8)

# Set the range for each axis.
# axis() function requires four values: the minimum and maximum values for the x-axis and the y-axis.
plt.axis([0, 1100, 0, 1100000])

plt.savefig('square_numbers_range_plot.png', bbox_inches='tight')
plt.show()
```
