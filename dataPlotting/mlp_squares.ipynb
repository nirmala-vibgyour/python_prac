```python
import matplotlib.pyplot as plt
# here, squares are plotted as the y-axis and x-axis: [0,1,2,3,4] (by default)
# so, the coordinates are (0,1), (1,4), (2,9), (3,16) and (4,25)
num_1 = [1, 2, 3, 4, 5]
num_2 = [1, 4, 9, 16, 25]
plt.plot(num_1, num_2, linewidth = 0.5)

# Set chart title and label axes.
plt.title("Simple plot", fontsize = 12)
plt.xlabel("num_1", fontsize = 9)
plt.ylabel("num_2", fontsize = 9)

# Set size of tick labels.
plt.tick_params(axis='both', labelsize=8)

plt.show()
plt.savefig('numbers_plot.png', bbox_inches='tight')
```
