```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("titanic.csv")
df
```

```python
df.info()
```

```python
# plots the dataframe
# float and int cols are plotted
# line plot of each cols with diff color
df.plot()
plt.show()
```

```python
# subplots, size of each, wheather or not common x and y axis
df.plot(subplots = True, figsize = (12, 15), sharex = False, sharey = False)
plt.show()
```

```python
# customised graph divisions
xticks = [ x for x in range(0, 901, 50)]
yticks = [ y for y in range(0, 81, 5)]
```

```python
df.age.plot(figsize = (10, 6), fontsize = 10, c = 'brown', linestyle = "-", xlim = (0, 900) , ylim = (0, 80), xticks = xticks, yticks = yticks, rot = 45)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger count", fontsize = 15)
plt.ylabel("Age", fontsize = 12)
plt.grid()
plt.show()
```

HISTOGRAM

```python
df.plot(kind = "hist", figsize = (12, 8), fontsize = 15)
plt.show()
```

```python
df.age.plot(kind = "hist", figsize = (12, 8), fontsize = 15, bins = 80)
plt.legend(loc = "best", fontsize = 15)
plt.show()
```

```python
df.age.hist(figsize = (12, 8), xlabelsize = 15, ylabelsize = 15, bins = 80, edgecolor='none', color='#496fad')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.hist(df.age.dropna(), bins = 80, density = True , cumulative = True, edgecolor='none', color='#496fad')
plt.xlim(-5, 85)
plt.ylim(0.0, 1.1)
plt.show()
```

```python
colors = [
    'b',    # Blue
    'g',    # Green
    'r',    # Red
    'c',    # Cyan
    'm',    # Magenta
    'y',    # Yellow
    'k',    # Black
    'w',    # White
    'orange',
    'purple',
    'brown',
    'pink',
    'gray',
    'olive',
    'teal',
    'navy',
    'maroon',
    'gold',
    'silver',
    'lightblue',
    'lightgreen',
    'lightcoral'
]
df1.plot(kind = "bar", figsize = (12, 8), fontsize = 12, rot = 45, color = "c" )
plt.show()
df1.shape
```
