Indexing
zero-based indexing : 0, 1, 2, 3, 4, 5, 6, ... for both row and col
negative indexing: ...., -5, -4, -3, -2, -1 for both row and col

iloc : postion based indexing and slicing

```python
import pandas as pd
df = pd.read_csv(r".\allFiles\Course_Materials_Part2\Coding_Exercises\summer.csv", index_col = "Athlete") 
# index_col: which col to be set as the index
df
```

```python
df.info() # the number of columns has reduced as Athelet col is used for the index
```

```python
df.iloc[456] 
type(df.iloc[456]) #iloc gives one row, here it gives the row 456
```

```python
df.iloc[[34,67,89, 90]] # the list is passed giving the index of four rows
```

```python
# rows in the range, here 45 is included but not 51
df.iloc[45:51]
```

```python
# first five rows, here it is rows 0, 1, 2, 3, 4
df.iloc[:5]
```

```python
# last five rows: -5, -4, -3, -2, -1
df.iloc[-5:]
```

```python
# df.iloc[row,col]
# df.iloc[0,4], it means first row and 4th column
# df.iloc[0,:3], it means first row and columns: 0, 1, 2 since 3 is not inclusive
# df.iloc[:,4], it means all rows column 4
```

```python
df.iloc[:,4].equals(df.Country)
```

```python
df.iloc[:,4].equals(df["Country"])
```

loc: label based indexing and slicing

```python
# uisng the index label instead of index number
df.loc["PERSAKIS, Ioannis"].equals(df.iloc[45])
```

```python
df.loc["PERSAKIS, Ioannis"]
```

```python
# df.loc[rows,cols]
df.loc["PERSAKIS, Ioannis", "Sport"]
# this is just an item
type(df.loc["PERSAKIS, Ioannis", "Sport"])
```

```python
df.loc["PERSAKIS, Ioannis", ["Sport", "Event", "Medal"]]
```

```python
# this is the single row's diff cols
type(df.loc["PERSAKIS, Ioannis", ["Sport", "Event", "Medal"]])
```

```python
df.loc[["PERSAKIS, Ioannis", "JANIKOWSKI, Damian"], ["Sport", "Event", "Medal"]]
```

```python
# this is the dataframe as there is more than one columm
type(df.loc[["PERSAKIS, Ioannis", "JANIKOWSKI, Damian"], ["Sport", "Event", "Medal"]])
```

```python
# this have all the rows for these three columns
df.loc[:, ["Sport", "Event", "Medal"]]
```
